```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(font_scale=1.3)
plt.rcParams['figure.dpi'] = 50
```

### Summary: Linear Vs Quadratic Vs Cubic

```python
x = np.arange(-3.5, 3.1, 0.1)

def f_linear(x):
    # ax + b
    return 3 * x + 4

def f_quadratic(x):
    # ax^2 + bx + c
    return 2 * x**2 + 3 * x + 2

def f_cubic(x):
    # ax^3 + bx^2 + cx + d
    return 3 * x**3 + 2 * x**2 + 3 * x + 1

sns.lineplot(x=x, y=f_linear(x), label='linear')
sns.lineplot(x=x, y=f_quadratic(x), label='quadratic')
sns.lineplot(x=x, y=f_cubic(x), label='cubic')
plt.ylim(-10, 10)
```

#### What is a Polynomial?

In mathematics, a polynomial is an expression consisting of indeterminates (also called variables) and coefficients, that involves only the operations of addition, subtraction, multiplication, and non-negative integer exponentiation of variables. An example of a polynomial of a single indeterminate x is x2 − 4x + 7. An example in three variables is x3 + 2xyz2 − yz + 1.

# Quadratic Graphs

$y=$ anything with $x^2$ (but no higher powers of x)

```python
x1 = np.arange(-3.2, 3.3, 0.1)
x2 = np.arange(-1.5, 3.6, 0.1)
x3 = np.arange(-4.0, 2.1, 0.1)

y1 = lambda x: x**2
y2 = lambda x: 3*x**2 - 6*x - 3
y3 = lambda x: -2*x**2 - 4*x + 3

fig, ax = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
sns.lineplot(x=x1, y=y1(x1), ax=ax[0])
sns.lineplot(x=x2, y=y2(x2), ax=ax[1])
sns.lineplot(x=x3, y=y3(x3), ax=ax[2])

plt.show()
```

### Graph Transformations

Let's start by defining a function f(x); we'll use the sin wave

```python
def f(x):
    return np.sin(x)

x = np.arange(-5.0, 5.1, 0.1)
```

#### Translation on the <B>y-axis</B>: $y = f(x) + a$

```python
sns.lineplot(x=x, y=f(x), label='f(x)')
sns.lineplot(x=x, y=f(x) + 1.5, label='f(x) + 1', linestyle='--')
```

#### Translation on the <B>x-axis</B>: $y = f(x - a)$

```python
sns.lineplot(x=x, y=f(x), label='f(x)')
sns.lineplot(x=x, y=f(x + 1), label='f(x + 1)', linestyle='--')
```

#### Reflations: $y = -f(x)$

```python
sns.lineplot(x=x, y=f(x), label='f(x)')
sns.lineplot(x=x, y=-f(x), label='-f(x)', linestyle='--')
```

#### Reflations: $y = f(-x)$

```python
def f(x):
    return 1 * np.power(x, 3) + 2 * np.power(x, 2) + 3*x

sns.lineplot(x=x, y=f(x), label='f(x)')
sns.lineplot(x=x, y=f(-x), label='f(-x)', linestyle='--')
plt.ylim(-10, 10)
```
