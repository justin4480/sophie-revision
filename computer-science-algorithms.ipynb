```python
chart = True
data = np.random.randint(0, 100, 100)

start = perf_counter()
sorted = False
while sorted == False:
    sorted = True
    for i in range(len(data)-1):
        t0 = data[i+0]
        t1 = data[i+1]
        if t1 < t0:
            data[i+0] = t1
            data[i+1] = t0
            sorted = False
    if chart:
        clear_output(wait=True)
        plt.figure(figsize=(18, 5))
        plt.bar(x=range(len(data)), height=data)
        plt.show()

if not chart:
    print(np.round(perf_counter() - start, 4), 'seconds')
```
